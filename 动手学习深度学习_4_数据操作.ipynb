```python
import os
 
os.makedirs(os.path.join('.','data1'),exist_ok=True)
data_file=os.path.join('.','data1','house_tiny.csv')
with open(data_file,'w') as f:
    f.write('NumRooms,Alley,Price\n')
    f.write('NA,Pave,127500\n')
    f.write('2,NA,106000\n')
    f.write('4,NA,178100\n')
    f.write('NA,NA,178100\n')
```

```python
# 如果没有安装pandas，只需取消对以下行的注释来安装pandas
#一般用pandas来读取csv文件 加载原始数据集
import pandas as pd

data = pd.read_csv(data_file)
print(data)
```

```python
# 处理缺失值 典型的方法包括 插值和删除 这里考虑插值
import pandas as pd

# 假设 data 已经定义并包含数据
inputs, outputs = data.iloc[:, 0:2], data.iloc[:, 2]

# 只对数值列进行插值
numeric_columns = inputs.select_dtypes(include='number').columns
inputs[numeric_columns] = inputs[numeric_columns].fillna(inputs[numeric_columns].mean())

print(inputs)
```

```python
# 对于inputs中的类别值或离散值，我们将“NaN”视为一个类别。 由于“巷子类型”（“Alley”）列只接受两种类型的类别值“Pave”和“NaN”， pandas可以自动将此列转换为两列“Alley_Pave”和“Alley_nan”。 巷子类型为“Pave”的行会将“Alley_Pave”的值设置为1，“Alley_nan”的值设置为0。 缺少巷子类型的行会将“Alley_Pave”和“Alley_nan”分别设置为0和1。
inputs = pd.get_dummies(inputs, dummy_na=True)
print(inputs)
```

```python
#转换为张量格式
import torch

X = torch.tensor(inputs.to_numpy(dtype=float))
y = torch.tensor(outputs.to_numpy(dtype=float))
X, y
#  NumRooms  Alley_Pave  Alley_nan
# 0       3.0        True      False
# 1       2.0       False       True
# 2       4.0       False       True
# 3       3.0       False       True
```
