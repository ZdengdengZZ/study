```python
#表示任意形状张量的元素和
A = torch.arange(20*2).reshape(2,5,4)
A.shape,A.sum
```

```python
#指定求和汇总张量的轴
a = A.sum(axis = 0)
a,a.shape
```

```python
#指定求和汇总张量的轴
a = A.sum(axis = 1)
a,a.shape
```

```python
#还可以求相关量的平均值
A.mean()
```

```python
#计算总和或均值时保持轴数不变
sum_A = A.sum(axis = 1,keepdims = True)
sum_A
```

```python
#通过广播机制将
A/sum_A
```

```python
#某个轴A元素的累积总和
A.cumsum(axis = 0)
```

```python
#点积是相同位置按元素乘积的和
y = torch.ones(4,dtype = torch.float32)
x = y
x,y,torch.dot(x,y)
```

```python
#与上面相同
torch.sum(x*y)
```

```python
A.shape,x.shape,torch.mv(A,x)
```

```python
B = torch.ones(4,3)
torch.mm(A,B) #执行m次矩阵向量积 并将结果拼接在一起 形成nxm矩阵
```

axis=0 去掉5只留下5

当keepdims = True的话将某一维度变为1 而不是直接去掉
